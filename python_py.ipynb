```python
#loops and control statement-->
#  break-->this statement break the loop
i=0  #intialisation
while i<5:
  i+=1 #increment
  if i==3:
    break
  else:
    print(i)
```

```python
#continue statement-->this statement skip the current iteration
i = 0
while i<5:
 i=i+1
 if i==3:
   continue
   print(i)
 else:
    print(i)
```

```python
#pass statement-->this statement pass the code
i=0
while i<5:
  i+=1
  if i==3:
    pass
    print(i)
  else:
    print(i)
```
